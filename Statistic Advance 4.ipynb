### Question 2
Q2: Differentiate between one-tailed and two-tailed tests.

A one-tailed test is a statistical test in which the critical region (the region where we reject the null hypothesis) is entirely contained within one tail of the sampling distribution. This type of test is used when the alternative hypothesis specifies a direction (e.g., greater than or less than). For example, if we want to test whether a new drug lowers blood pressure, we would use a one-tailed test with the critical region in the left tail of the sampling distribution.



A two-tailed test is a statistical test in which the critical region is split between the two tails of the sampling distribution. This type of test is used when the alternative hypothesis does not specify a direction. For example, if we want to test whether a new drug has any effect on blood pressure (either lowering or raising it), we would use a two-tailed test with critical regions in both tails of the sampling distribution.



__In General__

The main difference between one-tailed and two-tailed tests is the directionality of the alternative hypothesis and the location of the critical region(s).




### Question 3

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.


__Type 1 error__

Type 1 error occurs when the null hypothesis is rejected when it is actually true. This means that we conclude that there is a significant effect when in reality there is no effect. In addition, the probability of making a Type 1 error is denoted by the Greek letter alpha (Î±) and is also known as the significance level. 

For example, if a researcher conducts a study to test whether a new drug is effective in treating a disease and concludes that the drug is effective when it is actually not, then a Type 1 error has occurred.


__Type 2 error__

A Type 2 error occurs when the null hypothesis is not rejected when it is actually false. This means that we conclude that there is no significant effect when in reality there is an effect. The probability of making a Type 2 error is denoted by the Greek letter beta (Î²). 

For example, if a researcher conducts a study to test whether a new drug is effective in treating a disease and concludes that the drug is not effective when it actually is, then a Type 2 error has occurred.

__In general__

In summary, a Type 1 error is a false positive and a Type 2 error is a false negative.

Q4: Explain Bayes's theorem with an example.



QBayesâ€™ theorem is a mathematical formula that describes the relationship between two events and their probabilities. It is used to calculate the probability of an event occurring based on prior knowledge or evidence. The formula for Bayesâ€™ theorem is:

P(A|B) = (P(B|A) * P(A)) / P(B)

where P(A|B) is the probability of event A occurring given that event B has occurred, P(B|A) is the probability of event B occurring given that event A has occurred, P(A) is the prior probability of event A occurring, and P(B) is the total probability of event B occurring.

__Example__

To calculate the probability that a person has a certain disease given that they have tested positive for it before. Letâ€™s say that the prevalence of the disease in the population is 0.1% (i.e., P(Disease) = 0.001), the sensitivity of the test (i.e., the probability of testing positive given that the person has the disease) is 99% (i.e., P(Positive|Disease) = 0.99), and the specificity of the test (i.e., the probability of testing negative given that the person does not have the disease) is 95% (i.e., P(Negative|No Disease) = 0.95).

Using Bayesâ€™ theorem, we can calculate the probability that a person has the disease given that they have tested positive as follows:

P(Disease|Positive) = (P(Positive|Disease) * P(Disease)) / P(Positive) = (0.99 * 0.001) / ((0.99 * 0.001) + (1 - 0.95) * (1 - 0.001)) â‰ˆ 0.02

So, even though the person has tested positive for the disease, there is only about a 2% chance that they actually have it.

### Question 5

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

__Answer_

Confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. It is calculated from a sample of data and provides an estimate of the range in which the true population parameter is likely to lie.

The formula for calculating a confidence interval for a population mean depends on whether the population standard deviation is known or not. If the population standard deviation is known, then the formula for a confidence interval for the population mean is:

sample mean +/- (critical value) * (population standard deviation / sqrt(sample size))

If the population standard deviation is not known, then the formula for a confidence interval for the population mean is:

sample mean +/- (critical value) * (sample standard deviation / sqrt(sample size))

Thus, critical value is determined by the confidence level and can be found using a z-table or t-table depending on whether the population standard deviation is known or not.


__Example__

Suppose we have a random sample of 100 people and we want to estimate their average height with 95% confidence. The sample mean height is 170 cm and the sample standard deviation is 10 cm. Since the population standard deviation is not known, we use the formula for a confidence interval when the population standard deviation is not known:

170 +/- (1.96) * (10 / sqrt(100)) = (168.04, 171.96)

So, we are 95% confident that the true average height of the population lies between 168.04 cm and 171.96 cm.

Q8 What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

__Answer__

The margin of error is a measure of the precision of a confidence interval. It is equal to half the width of the entire confidence interval and represents the maximum amount by which the sample estimate is expected to differ from the true population parameter.

As I explianed in question 7, the sample size affects the margin of error because as the sample size increases, the standard error of the mean decreases. Resulting in margin of error decreasing as the sample size increases, resulting in a more precise estimate of the population parameter.

__Example__

Hereâ€™s an example to illustrate how a larger sample size would result in a smaller margin of error: Suppose we want to estimate the average height of a population with 95% confidence. We collect a random sample of 100 people and find that their average height is 170 cm with a standard deviation of 10 cm. The margin of error for this estimate would be approximately 1.96 * (10 / sqrt(100)) = 1.96.

Now suppose we collect a larger sample of 1000 people and find that their average height is also 170 cm with a standard deviation of 10 cm. The margin of error for this estimate would be approximately 1.96 * (10 / sqrt(1000)) = 0.62.

In conclusion, the larger sample size results in a smaller margin of error, indicating that our estimate is more precise.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

__Answer__

H0: The new drug is not significant at 95% : Âµ = 0
Ha: The new drug is significant at 95%:   Âµ != 0



we are using t-test (2 tails)

t_statistics = (sample mean - population mean) / (sample standard deviation / sqrt(sample size)) 

t_statistics = (6-0)/(2.5/âˆš(50))


```python
## code to solve this using the formular
def onetail_ttest(population_mean, sample_size, sample_mean, sample_std):
    import math  ## for sqrt
       
    ## t value calculated
    t_cal = (sample_mean-population_mean)/(sample_std/math.sqrt(sample_size))   
    return t_cal 

t_statistics = onetail_ttest(0, 50, 6, 2.5)
# print(t_statistics)
print(f"T_statistics: {t_statistics}\n")

def t_tabulated(df, alpha):
    from scipy.stats import t
    
    # degrees of freedom
        
    # one-tailed critical value
    critical_value = t.ppf(alpha, df)
    
    return critical_value

critical_value = t_tabulated(49, 0.95)

print(f"Critical Value: {critical_value}\n")

if t_statistics > critical_value:
    print("Reject the null hypothesis")
    print('''
    Since our calculated test statistic is greater than the critical value, 
    we reject the null hypothesis and conclude that there is significant 
    evidence at the 95% confidence level to suggest that the weight loss drug 
    is effective.
    ''')
else:
    print("Fail to reject the null hypothesis")
```
